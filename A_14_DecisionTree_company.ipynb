```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
cd=pd.read_csv('Company_Data.csv')
cd
```

```python
cd.head()
```

```python
cd.describe()
```

```python
#We will take high sales above at 7.49 beacause it is the average of sales
cd["sales"]=(cd["Sales"]>7.49)*1
```

```python
#We will create numerical variable for categorical data
label_encoder = preprocessing.LabelEncoder()
cd['ShelveLoc']= label_encoder.fit_transform(cd['ShelveLoc']) 
cd['Urban']= label_encoder.fit_transform(cd['Urban']) 
cd['US']= label_encoder.fit_transform(cd['US']) 
```

```python
cd
```

```python
cd.drop(["Sales"],axis=1,inplace=True)
```

```python
x=cd.iloc[:,0:10]
y=cd.iloc[:,10]
```

```python
cd.info()
```

```python
import seaborn as sns
sns.heatmap(cd.corr()>0.6, annot = True)
plt.show()
```

```python
# correlation matrix
sns.heatmap(cd.corr())
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#PLot the decision tree
tree.plot_tree(model);
```

```python
fn=['CompPrice','Income','Advertising','Population','Price','ShelveLoc','Age','Education','Urban','US','income']
cn=['YES','NO']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (5,5), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True); 
```

```python
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.feature_importances_
```

```python
y_pred = model.predict(x_test)
```

```python
#Here first column is for 'Good', & second is for 'Risky'
model.predict_proba(x_test)
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix, classification_report
cm= confusion_matrix(y_test, y_pred) 
cm
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True,  cmap = 'Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(model.score(x_test, y_test))
plt.title(all_sample_title, size = 15)
```

```python
# Predicting the values of test data
y_pred = model.predict(x_test)
print("Classification report - \n", classification_report(y_test,y_pred))
```

```python
train_accuracy = []
test_accuracy = []
for max_depth in range(1,11):
    model = DecisionTreeClassifier(criterion='entropy', max_depth= max_depth, random_state= 40)
    model.fit(x_train, y_train)
    train_accuracy.append(model.score(x_train, y_train))
    test_accuracy.append(model.score(x_test, y_test))
```

```python
accuracies = pd.DataFrame({'max_depth':range(1,11),'train_acc':train_accuracy, 'test_acc':test_accuracy})
accuracies
```

```python
# visulizing accuracy graph to look over max_depth
labels =['train_acc','test_acc']
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot()
plt.plot(accuracies['max_depth'],accuracies['train_acc'], marker = 'o')
plt.plot(accuracies['max_depth'], accuracies['test_acc'], marker = 'o')
ax.set_xticks(np.arange(1,11))
plt.legend(labels)
plt.xlabel('max_depth of the tree')
plt.ylabel('performance')
plt.show()
```
